## Source Data 

<!--- Replace the text below with a Markdown bullet list that defines the columns of the CSV file. Be sure to indicate the data type for each column. --->
- ID - Unique identifier for each person measured, integer
- Waist - Waist measurement for each person measured, integer
- Hip - Hip measurement for each person measured, integer
- Gender - Gender of each person measured, M or F

## Data Import

```python
# Goal: Extract the data from the file
import pandas as pd
w2h_df = pd.read_csv('w2h_data.csv')
w2h_df
# use proper pandas function to read data from a CSV file to a DataFrame

```

## Calculations

```python
# Goal: For each row of data calculate and store the w2h_ratio and shape.

# Step 1: Make sure the columns are in correct data types - refer to previous parts for the correct data types
w2h_df.dtypes


```

```python
# Step 2: calculate the 'W2H_Ratio' column using the 'Waist' and 'Hip' columns
# waist/hip
w2h_df['w2h_ratio'] = w2h_df['Waist']/w2h_df['Hip']
w2h_df
```

```python
# Step 3: Create the `Shape` column based on the values of the `W2H_Ratio` column - refer to previous parts for the logic
#Add a new column for 'Shape'
w2h_df['Shape'] = 'Pear'
w2h_df.loc[(w2h_df['Gender'] == 'M') & (w2h_df['w2h_ratio']>0.9) | (w2h_df['Gender'] == 'F') & (w2h_df['w2h_ratio']>0.8),'Shape']='Apple'
```
